## Main functions below

### Every time that any of the cells run training sessions it will automatically save the Q_table if not completely useless.
To further improve i could have made it so that i keep randomizing sizes and other values for more shots at a great q_table

```python
import numpy as np
import flappy_bird_gym
import matplotlib.pyplot as plt
import time
import pickle

# Discretize the observation space
def discretize_observation(observation, bins):
    discretized = list()
    for i in range(len(observation)):
        discretized.append(np.digitize(observation[i], bins[i]) - 1)
    return tuple(discretized)

def calculate_dynamic_reward(new_state, reward, distance_scale_factor):
    horizontal_distance, vertical_distance = new_state

    # Scale the vertical distance based on the horizontal distance
    # As the horizontal distance decreases, the impact of vertical distance increases
    scaled_vertical_distance = vertical_distance * (1 / (horizontal_distance + 1)) * distance_scale_factor

    # Calculate exponential punishment based on scaled vertical distance
    punishment = np.exp(np.abs(scaled_vertical_distance))

    # Adjust the reward: positive for passing a pipe, reduced by vertical misalignment
    modified_reward = reward - punishment
    return modified_reward

def train_agent(env, Q_table, bins, alpha, gamma, epsilon, total_episodes, distance_scale_factor):
    rewards_per_episode = []
    durations_per_episode = []  # List to store the duration of each episode

    for episode in range(total_episodes):
        current_state = discretize_observation(env.reset(), bins)
        total_reward = 0
        duration = 0

        done = False
        last_horizontal_distance = np.inf  # Initialize with a high value

        while not done:
            # Exploration-exploitation trade-off
            action = env.action_space.sample() if np.random.random() < epsilon else np.argmax(Q_table[current_state])
            
            # Take action and observe new state
            new_state, reward, done, info = env.step(action)
            new_state_discrete = discretize_observation(new_state, bins)
            
            # Check if the bird has cleared a pipe
            if new_state[0] > last_horizontal_distance:
                reward *= 2  # Double the reward for clearing a pipe

            # Update the last horizontal distance
            last_horizontal_distance = new_state[0]
            
            # Modify the reward based on dynamic scaling of vertical distance
            modified_reward = calculate_dynamic_reward(new_state, reward, distance_scale_factor)

            # Update Q-table
            old_value = Q_table[current_state + (action,)]
            next_max = np.max(Q_table[new_state_discrete])
            new_value = (1 - alpha) * old_value + alpha * (modified_reward + gamma * next_max)
            Q_table[current_state + (action,)] = new_value

            current_state = new_state_discrete
            total_reward += modified_reward
            duration += 1

        rewards_per_episode.append(total_reward)
        durations_per_episode.append(duration)  # Store duration of the episode

    
    # Q_table saver
    last_100 = sum(rewards_per_episode[-100:]) / min(len(rewards_per_episode), 100)
    average_reward = sum(rewards_per_episode) / len(rewards_per_episode) if rewards_per_episode else 0
    
    
    # Lets not save any garbage when we run different training sessions
    if last_100 >= 3:
        filename = f"Q_table_{average_reward:.2f}_LASTA_{last_100:.2f}.pkl"
    
        with open(filename, 'wb') as file:
            pickle.dump(Q_table, file)
    
    
    return Q_table, rewards_per_episode, durations_per_episode  


def plot_rewards(rewards_per_episode, alpha, gamma, epsilon):
    plt.plot(rewards_per_episode)
    plt.title(f'Training Progress (alpha={alpha}, gamma={gamma}, epsilon={epsilon})')
    plt.xlabel('Episodes')
    plt.ylabel('Total Reward')
    plt.show()

def plot_durations(durations, alpha, gamma, epsilon):
    plt.plot(durations)
    plt.title(f'Time Spent in the Air (alpha={alpha}, gamma={gamma}, epsilon={epsilon})')
    plt.xlabel('Episodes')
    plt.ylabel('Duration')
    plt.show()

# Play one game
def play_one_game_with_visualization(Q_table, bins):
    # Create a new environment with visualization enabled
    env = flappy_bird_gym.make("FlappyBird-v0", screen_size=(screen_width, screen_height), pipe_gap=150)
    
    current_state = discretize_observation(env.reset(), bins)
    done = False
    total_reward = 0

    while not done:
        action = np.argmax(Q_table[current_state])  # Exploit learned values
        new_state, reward, done, _ = env.step(action)
        print(new_state)
        current_state = discretize_observation(new_state, bins)
        total_reward += reward
        env.render()  # Render the game
        time.sleep(1 / 30)  # FPS

    print(f"Total reward from this game: {total_reward}")
```

## Below is a cell for running one training session with adjusted alpha, gamma, epsilon and adjusted total episodes.
Helpfull for building the "final" q_table that will "solve" flappy birds.

```python
# Environment setup
screen_width = 288  # Example width
screen_height = 512  # Example height
env = flappy_bird_gym.make("FlappyBird-v0", screen_size=(screen_width, screen_height), pipe_gap=150)

# "DEFAULT" Q-learning parameters
#alpha = 0.1  # Learning rate
#gamma = 0.9  # Discount factor
#epsilon = 0.1  # Exploration rate
#total_episodes = 5000  # Total episodes for training

# Q-learning parameters
alpha = 0.01  # Learning rate
gamma = 0.92  # Discount factor
epsilon = 0.01  # Exploration rate
total_episodes = 14000  # Total episodes for training

# Discretization bins
num_bins = 16
bins = [np.linspace(-0.5, 0.5, num_bins), np.linspace(-0.5, 0.5, num_bins)]

# Initialize Q-table
Q_table = np.zeros([num_bins, num_bins, env.action_space.n])

# Train the agent
distance_scale_factor = 0.2
Q_table, rewards_per_episode, durations_per_episode  = train_agent(env, Q_table, bins, alpha, gamma, epsilon, total_episodes, distance_scale_factor)

plot_rewards(rewards_per_episode, alpha, gamma, epsilon)
plot_durations(durations_per_episode, alpha, gamma, epsilon)
```
